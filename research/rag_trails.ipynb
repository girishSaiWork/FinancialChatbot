```python
import pandas as pd
```

```python
financial_data = pd.read_csv('../data/cleaned_data.csv')
financial_data.head()
```

```python
financial_data_filter = financial_data[['pre_text','post_text','question','answer']]
financial_data_filter.head()
```

```python
from langchain.agents.agent_types import AgentType
from langchain_experimental.agents.agent_toolkits import create_pandas_dataframe_agent
from langchain.llms import Ollama

agent = create_pandas_dataframe_agent(
    Ollama(model="llama3.1", temperature=0),  # Specify Ollama model
    financial_data_filter,
    verbose=True,
    allow_dangerous_code=True
)
```

```python
agent.invoke("what was the percentage change in the net cash from operating activities from 2008 to 2009?")
```
