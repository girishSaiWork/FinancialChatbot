```python
import pandas as pd
```

```python
financial_data = pd.read_csv('../data/cleaned_data.csv')
financial_data.head()
```

```python
financial_data_filter = financial_data[['pre_text','post_text','question','answer']]
financial_data_filter.head()
```

```python
from langchain.embeddings import OllamaEmbeddings
from langchain.vectorstores import FAISS

embeddings = OllamaEmbeddings(model='mxbai-embed-large')
vectorstore = FAISS.from_texts(financial_data_filter['pre_text'] + financial_data_filter['post_text'], embeddings)
```

```python
from langchain.chains import RetrievalQA
from langchain.llms import Ollama

llm = Ollama(model="llama")
qa_chain = RetrievalQA.from_chain_type(llm, retriever=vectorstore.as_retriever(), chain_type="stuff")

# User question
question = "What was the percentage change in the net cash?"

result = qa_chain.run(question)
print(result)
```
