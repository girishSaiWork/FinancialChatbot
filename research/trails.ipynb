```python
f_data['pre_text'] = f_data['pre_text'].apply(lambda x: ' '.join(x))
f_data['post_text'] = f_data['post_text'].apply(lambda x: ' '.join(x))
f_data.head(5)
```

```python
f_qa_expanded = pd.json_normalize(f_data['qa'])
f_qa_expanded.head()
```

```python
f_annotation_expanded = pd.json_normalize(f_data['annotation'])
f_annotation_expanded.head()
```

```python
full_df = pd.concat([f_data,f_qa_expanded,f_annotation_expanded], axis=1)
full_df.head()
```

```python
full_df.to_csv('../data/cleaned_data.csv')
```
